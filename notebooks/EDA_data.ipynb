```python
# change tipe data TotalCharges to numerik (Float)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
print(df[df['TotalCharges'].isnull()].to_string())
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(0)

print("Jumlah data yang kosong :")
print(df.isnull().sum())
```

```python
cat_cols = df.select_dtypes(include='object').columns.tolist()
for col in cat_cols:
    print(col, sorted(df[col].unique())[:10])
```

```python
for col in df.select_dtypes(include=object):
    df[col] = df[col].str.strip()

service_binary_cols = [
    'MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection',
    'TechSupport','StreamingTV','StreamingMovies'
]

for col in service_binary_cols:
    df[col] = df[col].replace({'No internet service':'No', 'No phone service':'No'})

df['InternetService'] = df['InternetService'].replace({'No internet service':'No'})
```
